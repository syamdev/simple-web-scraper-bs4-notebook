```python
# create function for searching job, then display data as dataframe using pandas

def search_job(job):
  source = requests.get('https://www.karir.com/search?q=' + job)
  soup = bs.BeautifulSoup(source.text, 'lxml')
  elements = soup.find_all('div', {'class': 'opportunity-box'})

  date = []
  job = []
  company = []
  salary = []
  location = []

  for element in elements:
    job.append(element.find('h4', {'class': 'tdd-function-name'}).get_text())
    company.append(element.find('div', {'class': 'tdd-company-name'}).get_text())
    date.append(element.find('time').get_text())
    salary.append(element.find('span', {'class': 'tdd-salary'}).get_text())
    location.append(element.find('span', {'class': 'tdd-location'}).get_text())
    

  data_df = {'date_posted': date,
        'job_function': job,
        'company': company,
        'salary': salary,
        'location': location}

  df = pd.DataFrame(data_df)

  return df
```

```python
# insert keyword with string (for more than one word: use '+' as space)
search_job('web+developer')
```
